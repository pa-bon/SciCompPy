### Exercise 2.1
Create a long positive integer. Find the number of zeros. Hint: change the number to a string. 

````python
#create large random integer
from random import randint
integer = randint(10**1000, 10**1001)
print(integer)

#create a variable to hold the number of zeros
n_zeros = 0

#iterate through the decimal expansion, upon encountering a zero update the counter
for c in str(integer):
    if c == '0':
        n_zeros += 1 

#report results     
print(f"\nThe inetegr provided has {n_zeros} zeros in it's decimal expansion")
````
