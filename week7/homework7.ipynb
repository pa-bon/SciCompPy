```python
from Vector import Vector
```

```python
def find_axis(v1, v2):
    '''Returns a unit vector perpendicular to given vectors'''
    try:
        a = v1.cross(v2)
        return a / a.length()   #__truediv__() method added to Vector class
    except ZeroDivisionError:
        raise ValueError("The vectors are paraller or at least one is zero")
```

```python
v = Vector(1,0,3)
w = Vector(-1,-2,3)

a = find_axis(v,w)
print(a)
print(a.length())
```
