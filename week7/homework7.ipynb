### Exercise 7.1
Create a function `find_axis(v1, v2)` which returns the unit vector `v3`, where `v3` is perpendicular to the vectors `v1` and `v2`. If the vectors `v1` and `v2` are parallel (or zero) then the function should raise an exception (ValueError) [Hint: `v1` and `v2` are parallel if the cross product `v1 × v2` is zero]. Vectors are instances of the `Vector` class from Homework 6. 

```python
from Vector import Vector
```

```python
def find_axis(v1, v2):
    '''Returns a unit vector perpendicular to given vectors'''
    try:
        a = v1.cross(v2)
        return a / a.length()   #__truediv__() method added to Vector class
    except ZeroDivisionError:
        raise ValueError("Dot product zero: the vectors are parallel or at least one is zero")
```

```python
v = Vector(1,0,3)
w = Vector(-1,-2,3)
u = Vector(2,0,6)

message1 = f'For vectors v ({v}) and w ({w}), the axis is {find_axis(v, w)} of lenght {find_axis(v,w).length()}'
message2 = f'For paraller vectors v ({v}) and u ({u}), the search for axis results in:'
print(message1+'\n'+message2)

find_axis(v, u)
```

### Excercise 7.2
Create the following infinite iterators:

(a) returning 0, 1, 0, 1, 0, 1, ...,

(b) returning randomly 0 or 1 on demand,

(c) returning 0, 1, 0, -1, 0, 1, 0, -1, ... 

```python
import itertools
from random import choice

# (a) returning 0, 1, 0, 1, 0, 1, ... -> a cycle of period 2

cycle = itertools.cycle([0,1])

a = list(itertools.islice(cycle, 10))
print(a)

# (b) returning randomly 0 or 1 on demand -> random.choice([0,1]) wraped into a generator

randomly = map(choice, itertools.repeat([0,1]))

b = list(itertools.islice(randomly, 10))
print(b)

# (c) returning 0, 1, 0, -1, 0, 1, 0, -1, ... -> a cycle of period 4

cycle2 = itertools.cycle([0,1,0,-1])

c = list(itertools.islice(cycle2, 10))
print(c)
```
