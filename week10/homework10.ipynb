### Exercise 10.2
Create `pd.Series` where index contains days of the current month (`numpy.datetime64` instances or `pandas.DatetimeIndex`) and data contains some numbers (temperatures at noon, currency rates, ...).

````python
import pandas as pd
import numpy as np
````

````python
dates = np.arange(np.datetime64('2026-05-01'), np.datetime64('2026-06-01'))
lucky_n = np.random.randint(0, 10, len(dates))

lucky_numbers = pd.Series(data = lucky_n, index=dates, name='Lucky Numbers')

print(lucky_numbers)
````
