### Exercise 4.1

Let `p=(x,y)` be a point in a plane. Define the following functions using lambda:

1. a test if p is in unit (filled) circle,
2. a test if the coordinates of p are positive,
3. a sorting key (y decreasing, x increasing),
4. a sorting key (the sum |x|+|y|). 

```python
from random import randrange

#preparing the list of points to work on
mes1='The list of points is:'

points = []
for n in range(10):
    points.append((randrange(-10, 10, 1)/10, randrange(-10, 10, 2)/10))
    mes1 = mes1 + '\n' + str(points[n])

print(mes1)
```

```python
#Ad 1
is_in_unit = lambda p : p[0]*p[0] + p[1]*p[1] <= 1

in_unit = list(filter(is_in_unit, points))

mes2 = 'The points from the list within the unit circle:'
for p in in_unit:
    mes2 = mes2 + '\n' + str(p)
    
print(mes2)
```

```python
#Ad 2
positive_coordinates = lambda p : p[0] > 0.0 and  p[1] > 0.0

pos_coord = list(filter(positive_coordinates, points))

mes3 = 'The points from list with both coordinates positive:'
for p in pos_coord:
    mes3 = mes3 + '\n' + str(p)

print(mes3)
```

```python
#Ad 3
sort_yde_xin = lambda p : (-p[1], p[0])
points.sort(key=sort_yde_xin)

mes4='The list sorted by y decrising, x incrising:'
for p in points:
    mes4 = mes4 + '\n' + str(p)
print(mes4)
```

```python
#Ad 4
sort_abs = lambda p : abs(p[0]) + abs(p[1])
points.sort(key=sort_abs)
mes5='The list sorted by the xalue of |x| + |y|:'
for p in points:
    mes5 = mes5 + '\n' + str(p)
print(mes5)
```

### Exercise 4.2

Reversing a part of a list in place:

 `reverse_range(L, left, right)`
 
  the right index is included. Write iterative and recursive versions. 

```python
def reverse_range(L, left, right):
    '''Reverses part of the list from index "left" to "right" in place in iterative manner'''
    for (n, i) in enumerate(L):                                 #go through items
        if n >= left and n < ((left+right)/2):                  #if index of item is in first half of the range
            L[n], L[right-(n-left)] = L[right-(n-left)], i      #switch places of the item with its counterpart

L = list(range(10))

reverse_range(L, 3, 6, )
print(L)
```

```python
def reverse_range_recursive(L, left, right):
    '''Reverses part of the list from index "left" to "right" in place recursive manner'''
    L[left], L[right] = L[right], L[left]               #switch places of the first and last item in range
    if right-left > 1:                                  #unless the range is of 2 or less elements
        reverse_range_recursive(L, left+1, right-1)     #reapat with narrowed range

L = list(range(10))

reverse_range_recursive(L, 3, 6)
print(L)
```

### Exercise 4.3

Create the following infinite generators:
1. `iter_even()`, producing even numbers (0, 2, 4, ...),
2. `iter_odd()`, producing odd numbers (1, 3, 5, ...),
3. `iter_power(k)`, producing powers of k (1, k, k\*k, k\*k\*k, ...).

```python
def iter_even():
    '''Generates even numbers'''
    k = 0
    while True:
        yield 2*k
        k += 1

def iter_odd():
    '''Generates odd numbers'''
    k = 0
    while True:
        yield 2*k+1
        k += 1

def iter_power(k):
    '''Generates powers of k'''
    n=0
    while True:
        yield k**n
        n += 1

#preparing the headings of the table with results
results=('The iterators give flowing results:\n\n'+
    '|'.join(['', 'Lp.'.center(5), 'iter_even()'.center(15), 'iter_even()'.center(15), 'iter_power(k=3)'.center(25), ''])+
    '\n'+('-'*65)+'\n')

#initiating the generators
ie = iter_even()
io = iter_odd()
ip = iter_power(3)

#getting results
for i in range(21):
    e = next(ie)
    o = next(io)
    p = next(ip)
    results = results + '|'.join(['', str(i).center(5), str(e).center(15), str(o).center(15), str(p).center(25), '']) + '\n'

#rapporting results
print(results)
```
