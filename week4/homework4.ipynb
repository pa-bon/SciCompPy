### Exercise 4.1

Let `p=(x,y)` be a point in a plane. Define the following functions using lambda:

1. a test if p is in unit (filled) circle,
2. a test if the coordinates of p are positive,
3. a sorting key (y decreasing, x increasing),
4. a sorting key (the sum |x|+|y|). 

```python
from random import randrange

#preparing the list of points to work on
mes1='The list of points is:'

points = []
for n in range(10):
    points.append((randrange(-10, 10, 1)/10, randrange(-10, 10, 2)/10))
    mes1 = mes1 + '\n' + str(points[n])

print(mes1)
```

```python
#Ad 1
is_in_unit = lambda p : p[0]*p[0] + p[1]*p[1] <= 1
print(f'The first point is in unit circle: {is_in_unit(points[1])}')
```

```python
#Ad 2
positive_coordinates = lambda p : p[0] > 0 and  p[1] > 0
print(f'The second point has both coordinates positive: {is_in_unit(points[1])}')
```

```python
#Ad 3
sort_yde_xin = lambda p : (-p[1], p[0])
points.sort(key=sort_yde_xin)

mes2='The list sorted by y decrising, x incrising:'
for p in points:
    mes2 = mes2 + '\n' + str(p)
print(mes2)
```

```python
#Ad 4
sort_abs = lambda p : abs(p[0]) + abs(p[1])
points.sort(key=sort_abs)
mes3='The list sorted by the xalue of |x| + |y|:'
for p in points:
    mes3 = mes3 + '\n' + str(p)
print(mes3)
```

### Exercise 4.2

Reversing a part of a list in place, `reverse_range(L, left, right)`, the right index is included. Write iterative and recursive versions. 

```python
L = list(range(10))

def reverse_range(L, left, right):
    new_L = L[:left]
    new_L = new_L + L[left:right+1][::-1]
    new_L = new_L + L[right+1:]
    return new_L

print(reverse_range(L, 2, 4))
```

```python
L = list(range(10))

def list_reversor(L):
    reversed_list = []
    for i in L:
        reversed_list = [i] + reversed_list
    return reversed_list

print(list_reversor(L))

def list_reversor2(L):
    if len(L) == 2:
        [a, b] = L
        return [b, a]
    reversed_list = [L[-1]] + list_reversor(L[:-1])
    return reversed_list

print(list_reversor2(L))
```
