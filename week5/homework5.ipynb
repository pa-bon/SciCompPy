### Exercise 5.1 
For a given directory (`top`) find the number of bytes taken by PDF files in the directory tree (".pdf" extensions). The code should be in the function `find_pdf_size(top)`. In order to test the current directory we run `find_pdf_size(".")`. 

```python
import os

def find_pdf_size(top):
    '''Returns the combined size of PDF files within the given directory tree'''
    pdf_size = 0 #counter

    #go through directory tree
    for root, dirs, files in os.walk(top):
        for name in files:
            if name.lower().endswith(".pdf"):                       #for pdf files
                filepath = os.path.join(root, name)                 #recontruct filepath
                pdf_size = pdf_size + os.path.getsize(filepath)     #check size, update counter
    
    #return results
    return pdf_size

top = 'dir1'

find_pdf_size(top)
```

### Exercise 5.2
Create the function `print_working_days(date1, date2)`, where `date1` and `date2` are strings of the form 'YYYY-MM-DD'. The function prints dates of working days (from Monday to Friday) in the given range, `date2` is excluded. 

```python
import datetime

def print_working_days(date1, date2):
    '''Prints dates of working days since date1 to date2 (exclusive). Both dates must be provided in 'YYYY-MM-DD' format'''
    
    #decode dates
    d1 = datetime.date.fromisoformat(date1)
    d2 = datetime.date.fromisoformat(date2)
    
    dn = d1          #current day holder
    
    #go throu all days in range, print dates of working days
    while dn < d2:
        if datetime.date.weekday(dn) < 5:
            print(datetime.date.isoformat(dn))
        dn = dn + datetime.timedelta(1)

print_working_days('2026-04-15', '2026-04-30')
```

### Exercise 5.3 
Create the generator `random_walk(start)` for a 1D random walker. If a position at a certain moment is x, then the next position can be x+1 or x-1 with equal probabilities. Find the final position after 100 moves (start=0). Repeat experiments. 

```python
from random import randrange

#define generator function
def random_walk(start=0):
    '''A 1D random walk generator'''
    position = start
    while True:
        position = position + randrange(-1, 2, 2)
        yield position

results = {}    #simulation resluts: how many times each final position was reached?

#do 100 symulation of the random walk with 100 steps
for sym in range(100):
    
    #new generator for each symulation
    rw = random_walk()  
    
    #do 100 steps
    for step in range(100):
        position = next(rw)
    
    #record final position; if alredy exist, update the counter of it's ocurrences
    if not position in results.keys():
        results[position] = 1
    else:
        results[position] += 1

#sort results
final_positions = list(results.keys())
final_positions.sort()

#generate text-based histogram
histogram = 'The results of 100 random 1D walks with 100 steps\n'

for result in final_positions:
    histogram = histogram + (
        str(result).rjust(4) + ' : ' + '▮'*results[result] + '\n'
    )

print(histogram)
```
